### Smoker vs Cancer patients

```python
sns.countplot(x='smoker', hue='Cancer history', data=joined_df)
```

```python
number_of_smokers = medical_examinations[medical_examinations['smoker'] == 1]['Customer ID'].nunique()
cancer = medical_examinations[(medical_examinations['Cancer history'] == 1) & (medical_examinations['smoker'] == 1)]['Customer ID'].nunique()
print('Total Number of people who smoke: ', number_of_smokers)
print('smokers who have cancer: ', cancer)
print('ratio: ', 1.0 * cancer / number_of_smokers)
```

```python
number_of_nonsmokers = medical_examinations[medical_examinations['smoker'] == 0]['Customer ID'].nunique()
cancer = medical_examinations[(medical_examinations['Cancer history'] == 1) & (medical_examinations['smoker'] == 0)]['Customer ID'].nunique()
print('Total Number of people who do not smoke: ', number_of_nonsmokers)
print('smokers who have cancer: ', cancer)
print('ratio: ', 1.0 * cancer / number_of_nonsmokers)
```

### Heart issues for differnt categories of HBA1C

```python
sns.countplot(x='remark_HBA1C', hue='Heart Issues', data=medical_examinations)
```

```python
number_of_borderline = medical_examinations[medical_examinations['remark_HBA1C'] == 'Borderline']['Customer ID'].nunique()
have_heart_issues = medical_examinations[(medical_examinations['Heart Issues'] == 1) & (medical_examinations['remark_HBA1C'] == 'Borderline')]['Customer ID'].nunique()
print('Total Number of borderline people: ', number_of_borderline)
print('Number of people who have heart issues: ', have_heart_issues)
print('ratio: ', 1.0 * have_heart_issues / number_of_borderline)
```

```python
number_of_critical = medical_examinations[medical_examinations['remark_HBA1C'] == 'Critical']['Customer ID'].nunique()
have_heart_issues = medical_examinations[(medical_examinations['Heart Issues'] == 1) & (medical_examinations['remark_HBA1C'] == 'Critical')]['Customer ID'].nunique()
print('Total Number of critical people: ', number_of_critical)
print('Number of people who have heart issues: ', have_heart_issues)
print('ratio: ', 1.0 * have_heart_issues / number_of_critical)
```

```python
number_of_Excellent = medical_examinations[medical_examinations['remark_HBA1C'] == 'Excellent']['Customer ID'].nunique()
have_heart_issues = medical_examinations[(medical_examinations['Heart Issues'] == 1) & (medical_examinations['remark_HBA1C'] == 'Excellent')]['Customer ID'].nunique()
print('Total Number of borderline people: ', number_of_borderline)
print('Number of people who have heart issues: ', have_heart_issues)
print('ratio: ', 1.0 * have_heart_issues / number_of_Excellent)
```
